```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('order_data.csv')
```

```python
df.head()
```

```python
categories = ["Electronics","Clothing","Home","Book"]

for i,cat in enumerate(categories):
    print("Category:", cat, " Index number:", i)
```

```python
flags = []

for rev in df["revenue"]:
    if rev >= 200:
        flags.append("large")
    else:
        flags.append("small")
```

```python
df["order_size"] = flags
df
```

```python
# Uses numpy 'Where' method
df["order_size_np"] = np.where(
    df["revenue"] >= 200,
    "large",
    "small"
)
df
```
