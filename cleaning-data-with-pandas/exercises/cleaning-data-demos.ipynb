```python
import pandas as pd
import numpy as np
```

```python
students = pd.read_csv("C:/Users/fpose/Launchcode/data-analysis-projects/cleaning-data-with-pandas/exercises/student_records_dirty.csv")
students["course_grade"] = students["course_grade"].replace("\"N/A\"", np.nan)
students
```

```python
students = students.drop_duplicates()
```

```python
students["first_name"] = students["first_name"].str.strip().str.upper()
students["last_name"] = students["last_name"].str.strip().str.upper()
```

```python
students["major"] = students["major"].str.strip().str.capitalize()
```

```python
students["major"] = students["major"].replace({
    "Comp sci" : "Computer Science",
    "Cs" : "Computer Science",
    "Computer science" : "Computer Science",
    "Bio" : "Biology",
    "Psych" : "Psychology"
})
```

```python
students["major"].value_counts()
```

```python
students["course_grade"].value_counts()
```

```python
# students["course_grade"] = students.loc[students["course_grade"] == " ninety", "course_grade"] = 90
students["course_grade"] = students["course_grade"].replace({
    " ninety ":90,
    " 85":85
})
students["course_grade"].value_counts()
```

```python
students["age"].describe()
```

```python
# Replacing bad outliers with NaN values so it's not included in analysis
students.loc[students["age"] < 5, "age"] = np.nan
students.loc[students["age"] > 120, "age"] = np.nan
students["age"].describe()
```

```python
students["full_name"] = students["first_name"] + " " + students["last_name"]
students
```

```python
students["age_group"] = pd.cut(
    students["age"],
    bins = [0,25,40,65], 
    labels = ["Young Adults", "Working Adults", "Retired"]
)
students["age_group"].value_counts()
```

```python
students["enrollment_date"] = pd.to_datetime(students["enrollment_date"])
students["enrollment_date"]
```
