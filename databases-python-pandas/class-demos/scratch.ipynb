```python
city_standards = {
    'LA': 'Los Angeles',
    'NYC': 'New York'
}

customers['city'] = customers['city'].replace(city_standards)
customers['city'] = customers['city'].str.title()
customers['city'].value_counts()
```

```python
# APPLY FUNCTION EXAMPLE
#technically don't need it here but for practice
customers['clean_last_name'] = customers['last_name'].apply(
    lambda last_name : # use function on every value
    last_name.strip() # strip function for each
)
```

```python
#another, row-wise
order_items["total_price"] = order_items["quantity"] * order_items["unit_price"]
```

```python
# Apply called on entire dataframe - 
# Limits it with axis = 1, telling it to go row by row
order_items["total_price_apply"] = order_items.apply(
    lambda row:
    row["quantity"] * row["unit_price"], axis = 1
)
```

```python
order_items["order_items"] = order_items.apply(
    lambda row: (
        "Large Expensive Order" if row["quantity"] >= 3 and row["unit_price"] > 50 
        else "Bulk Order" if row["quantity"] >= 3
        else "Standard Order"
    ), axis = 1
)
```
