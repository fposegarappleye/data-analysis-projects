```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('fitness_data.csv')
df.head()
```

```python
plt.plot(df["steps"]) # Creates the basic plot off the data
plt.title("Raw Steps Data")
plt.ylabel("Steps")
plt.xlabel("Row of data")
plt.show() # NB that this it's critical to 'show' 
```

```python
plt.figure(figsize = (15,8))
plt.plot(df["date"], df["steps"])
plt.plot("Number of Steps per Day")

plt.xlabel("Date")
plt.xticks(rotation = 45)
plt.tight_layout()

plt.ylabel("Number of Steps")

plt.show()
```

```python
plt.figure(figsize = (15, 5))
plt.plot(df["date"], df["steps"], color = 'crimson', linewidth = .4)
plt.title("Resting Heart Rate")
```

```python
plt.figure(figsize=(6,6))
plt.scatter(df["steps"], df["calories_burned"], alpha=0.6)
plt.title("Steps versus calories burned", fontsize = 20)
plt.xlabel("Steps taken", fontsize = 15)
plt.ylabel("Calories Burned", fontsize = 15)
plt.show()
```

```python
df.head()
```

```python
plt.figure(figsize = (6,4))
colors = {
    "Low": "lightblue",
    "Moderate": "green",
    "High": "orange"
}

for level, subset in df.groupby("activity_level"):
    plt.scatter(subset["steps"], subset["calories_burned"],
    label = level,
    color = colors[level],
    alpha = 0.6
    )

plt.title("Steps vs Calories - by Activity Level")
plt.xlabel("Steps")
plt.ylabel("Calories")
plt.legend()
```

```python
plt.hist(df["steps"], bins = 20, edgecolor = "white")
plt.title("Distribution of Daily Steps", fontsize = 20)
plt.xlabel("Steps taken")
plt.ylabel("Frequency (days)")
```

```python
fig, axes = plt.subplots(1,2)

axes[0].hist(df["sleep_hours"], bins = 10, color = "teal", edgecolor = 'white')
axes[0].set_title("Sleep Hours Distribution")

axes[1].hist(df["resting_hr"], bins = 10, color = "crimson", edgecolor = 'white')
axes[0].set_title("Resting Heart Rate (bpm) Distribution")


plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2,2, figsize = (6.5,6.5))

# Histogram 1 (top left)
axes[0,0].hist(df["steps"], bins = 20, color = "skyblue", edgecolor = "white")
axes[0,0].set_title("Daily Steps")

# Histogram 2 (top right)
axes[0,1].hist(df["calories_burned"], bins = 20, color = "orange", edgecolor = "white")
axes[0,1].set_title("Daily Steps")

# Histogram 2 (bottom left)
axes[1,0].hist(df["sleep_hours"], bins = 20, color = "green", edgecolor = "white")
axes[1,0].set_title("Hours of Sleep")

# Histogram 2 (bottom right)
axes[1,1].hist(df["resting_hr"], bins = 20, color = "darkred", edgecolor = "white")
axes[1,1].set_title("Resting Heart Rate")
```

```python
activity_counts = df["activity_level"].value_counts()
```

```python
plt.bar(activity_counts.index, activity_counts.values, color = 'pink')
plt.title('Acitivty Level Counts')
plt.xlabel("Activity Level")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize = (8,5))
df.boxplot(column = "sleep_hours", by="activity_level")
plt.title("Sleep Hours by Activity Level")
plt.suptitle("") # Supertitle
plt.xlabel("Activity Level")
plt.ylabel("Hours of Sleep")
plt.show()
```

```python
plt.figure(figsize = (8,5))
df.boxplot(column = "resting_hr", by="activity_level")
plt.title("Resting Heart Rate by Activity Level")
plt.suptitle("") # Supertitle
plt.xlabel("Resting Heart Rate")
plt.ylabel("Number of Hours of Sleep")
plt.show()
```

```python
fig, axes = plt.subplots(2,1, figsize = (10,8), sharex = True)

axes[0].plot(df["date"], df["resting_hr"], color="darkred")
axes[0].set_title("Resting Heart Rate Over Time")

axes[1].plot(df["date"], df["sleep_hours"], color="blue")
axes[1].set_title("Resting Heart Rate Over Time")
```

```python
fig, axes = plt.subplots(1,3, figsize = (15,5)) 
# (1,3) is 1 row, 3 plots

levels = ["Low", "Moderate", "High"]
colors = ["skyblue", "limegreen", "orange"]

# zip() maps them together into a dictionary, like a zipper, "zips" them together
for ax, level, color in zip(axes, levels, colors):
    subset = df[df["activity_level"] == level] # Divides it into levels by the zip function
    ax.scatter(subset["steps"], subset["calories_burned"], alpha = 0.6, color = color) # creates scatter plot
    ax.set_title(f"{levels} Activity") 
    ax.set_xlabel("Steps")
    ax.set_ylabel("Calories Burned")

plt.tight_layout()
plt.show()
```
